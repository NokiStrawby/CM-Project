```
kList = ["1/5", "2/5", "3/5", "4/5"]
errDiffList = list();

for dirPath in imgDir:
  imgPath = "selection/" + dirPath + "/out/";
  imgs = os.listdir(imgPath);
  imgList = [ img for img in imgs if (img.endswith(".csv") and not "opterrs" in img) ]
  for img in imgList:
    errList = list()
    optErrList = list()
    with open("selection/" + dirPath + "/out/" + img, newline='') as csvImg:
      csvReader = csv.reader(csvImg, delimiter=',')
      for row in csvReader:
        if row[0] != 'k':
          errList.append(float(row[1]));
      errList.reverse();
    errOptFile = (img.split("."))[0];
    with open("selection/" + dirPath + "/out/" + errOptFile + "_opterrs.csv", newline='') as csvErrOpt:
      csvReader = csv.reader(csvErrOpt, delimiter=',')
      for row in csvReader:
        if row[0] != 'opt_err':
          optErrList.append(float(row[0]));
      optErrList.reverse();
    diff = list(map(sub, errList, optErrList));
    errDiffList.append(diff);
    plt.figure();
    plt.title(img);
    plt.plot(kList, diff);
```

```
import pandas as pd

ks = ["1/5", "2/5", "3/5", "4/5"];

def plotErrs(df_err, frame_name):
  errs = list(df_err["err"]);
  errs.reverse();
  plt.figure();
  plt.title('Frame ' + frame_name);
  plt.plot(ks, errs);

def plotErrDiff(df_err, df_opt_err, frame_name):
  errs = list(df_err["err"]);
  errs.reverse();
  opt_errs = list(df_opt_err["opt_err"]);
  opt_errs.reverse();
  diff = list(map(sub, errs, opt_errs));
  plt.figure();
  plt.title('Frame ' + frame_name + ' difference with opt error');
  plt.plot(ks, diff);
  return diff;
  

#SEPARARE IN CELLE DIVERSE!!!!!
imgDir = os.listdir("selection/");
errLists = list();
for dirPath in imgDir:
  imgPath = "selection/" + dirPath + "/out/";
  imgs = os.listdir(imgPath);
  err_dfs = {img : pd.read_csv(imgPath + img) for img in imgs if not "opterrs" in img};
  opt_err_dfs = {img.replace("_opterrs", '') : pd.read_csv(imgPath + img) for img in imgs if "opterrs" in img};
  diff_errs = list();
  for df_key in err_dfs.keys():
    plotErrs(err_dfs[df_key], df_key.replace(".csv", ''));
    diff_errs.append(plotErrDiff(err_dfs[df_key], opt_err_dfs[df_key], df_key.replace(".csv", '')));

  plt.figure();
  for diff in diff_errs:
    plt.plot(ks, diff);

  

#media diff per ogni k per tutte le immagini di ogni categoria, plottare quello 
```
